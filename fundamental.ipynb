### Importing all the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing dataset

```python
dataset = pd.read_csv('fundamentals.csv')
```

```python
dataset.head()
```

```python
dataset.columns
```

#### 1. Selecting columns  ‘Ticker Symbol’, ‘Total Revenue’ and ‘Earnings Before Tax’


```python
fundamental = dataset[['Ticker Symbol','Total Revenue', 'Earnings Before Tax']]
```

```python
fundamental.columns
```

```python
fundamental.head(30)
```

```python
fundamental.shape
```

##### Here I did not find 'GOOG' data in Ticker Symbol so I used 'GPS' data instead of 'GOOG' data

```python

gps = fundamental[fundamental['Ticker Symbol'] == 'GPS']
gps
                                                    
```

```python
gps.shape
```

```python
aapl = fundamental[fundamental['Ticker Symbol'] == 'AAPL']
aapl
```

```python
aapl.shape
```

#### 2. Here we select two companies 's ticker symbol i.e 'GPS' and 'AAPL' by concatenating  both ticker symbol.

```python
concat = pd.concat([aapl,gps])
```

```python
concat.head(10)
```

#### 3. Earning Per share of 'AAPL' and 'GPS'


```python
concat['Total Equity'] = dataset[['Total Equity']]
concat.head(10)
```

```python
concat['Earning Per Share'] = concat['Earnings Before Tax'] / concat['Total Equity']
concat
```

```python
sns.scatterplot(x = 'Ticker Symbol', y = 'Earning Per Share', data = concat , hue = 'Ticker Symbol')
plt.title('Scatterplot Visualization of Earning Per Share')
plt.show()
```

```python
sns.pairplot(concat)
plt.show()
```

##### From this visualization we can conclude that 'GPS' has highest Earning Per Share than 'APPL'

#### 4. Finding 'Total Revenue' bigger and lesser  1B  and between 1B to 10B of top ten companies

##### Total Revenue bigger than  1B of top ten companies

```python
bigger = fundamental[fundamental['Total Revenue'] > 1000000000].head(10)
bigger
```

```python
sns.scatterplot(x = 'Ticker Symbol', y = 'Total Revenue',data = bigger, hue = 'Ticker Symbol')
plt.title('Scatterplot Visualization of total revenue of top ten companies')
plt.show()
```

```python
sns.pairplot(bigger)
```

#### From this visualization we can conclude that APPL has highest revenue whereas AAP has lowest revenue)

#### Total Revenue Lesser than 1B of top ten companies

```python
lesser = fundamental[fundamental['Total Revenue'] < 1000000000].head(10)
lesser
```

```python
sns.jointplot(x = 'Earnings Before Tax', y = 'Total Revenue', data = lesser)
plt.show()
```

```python
sns.scatterplot(x = 'Ticker Symbol', y = 'Total Revenue', data = lesser, hue = 'Ticker Symbol')
plt.title('Scatterplot Visualization of Total Revenue')
plt.grid(True)
plt.show()
```

#### From this visualization we can conclude that 'AIV' has highest and 'COTY' has lowest total revenue.

```python
between = fundamental[fundamental['Total Revenue'].between (1000000000 , 10000000000, inclusive = True )].head(10)
between
```

```python
sns.scatterplot(x = 'Ticker Symbol', y = 'Total Revenue',data = between, hue = 'Ticker Symbol')
plt.title("Scatter plot visualization of total revenue")
plt.grid(True)
plt.show()
```

#### From this visualization we can conclude that 'AAP' has highest and 'ADI' has lowest Total Revenue.
